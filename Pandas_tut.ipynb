```python

d1 = {'Name': ['Pankaj', 'Meghna', 'Pankaj', np.nan],
      'ID': [1, 2, 3, np.nan],
      'Salary': [100, 200, np.nan, np.nan],
      'Role': ['CEO', None, pd.NaT, np.nan]}

df = pd.DataFrame(d1)
```

```python
df
```

```python
df.dropna() # drops row having any null
```

```python
df = df.dropna(how ='all') # drops row having all null
```

```python
df.drop_duplicates('Name')
```

```python
df.drop_duplicates(subset='Name', keep='first') # keep= 'first'/'last'/False
```
